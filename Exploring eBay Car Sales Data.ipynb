3) Convert columns 'Price' and 'odometer' to numeric dtype.

```python
autos['price'] = autos['price'].str.replace('$', '')
autos['price'] = autos['price'].str.replace(',', '')
autos['price'] = autos['price'].astype(int)
```

```python
autos['odometer'] = autos['odometer'].str.replace('km', '')
autos['odometer'] = autos['odometer'].str.replace(',', '')
autos['odometer'] = autos['odometer'].astype(int)
autos.rename({'odometer': 'odometer_km'}, axis=1, inplace=True)
```

7) Exploring Price by Brand

-Selecting the 20 most published brands
-Getting the mean price of the 20 most published brands

```python
top20brands = autos['brand'].value_counts().head(20).index
print(top20brands)
```

```python
mean_price_of_brands = {}

for brand in top20brands:
    selected_brand = autos[autos['brand'] == brand]
    mean_price = selected_brand['price'].mean()
    mean_price_of_brands[brand] = int(mean_price)
    
print(mean_price_of_brands)
```

-Audi, BMW and Mercedes Benz are the most expensive


```python
mean_mileage_of_brands = {}

for brand in top20brands:
    selected_brand = autos[autos['brand'] == brand]
    mean_mileage = selected_brand['odometer_km'].mean()
    mean_mileage_of_brands[brand] = int(mean_mileage)
    
print(mean_mileage_of_brands)
```

-The dictionaries will be converted to series objects.

-Then converted to dateframe objects.

```python
mpb_series = pd.Series(mean_price_of_brands)
mmb_series = pd.Series(mean_mileage_of_brands)
```

```python
brand_analysis = pd.DataFrame(mpb_series, columns=['mean_price'])
brand_analysis
```

```python
brand_analysis.insert(1, 'mean_km', mmb_series)
brand_analysis
```
